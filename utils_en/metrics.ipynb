```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = '../model/training_metrics.csv'
data = pd.read_csv(csv_file)
# Read the CSV file that records training loss

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(data['epoch'], data['loss'], color='red', label='loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax1.set_title('Training Loss')
# Create a plot with Epoch on the X-axis and Loss on the Y-axis
# The curve label and tick marks are red, and set the title of the chart

plt.savefig('../training_metrics.png')
plt.show()
```
